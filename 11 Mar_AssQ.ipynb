**Q1. What is the difference between a t-test and a z-test? Provide an example scenario where you would use each type of test.**

**Ans:**
    
Both t-tests and z-tests are statistical hypothesis tests used to determine whether a sample is likely to have come from a certain population with a known or assumed distribution. However, the main difference between t-tests and z-tests is that t-tests are used when the sample size is small or when the population standard deviation is unknown, while z-tests are used when the sample size is large (usually greater than 30) or when the population standard deviation is known.

Here are example scenarios where each test might be appropriate:

T-Test: Suppose you are a researcher who is interested in comparing the average test scores of two groups of students, Group A and Group B. You have collected a random sample of 20 students from each group, and you want to know if the difference in the means of the two groups is statistically significant. Since the sample size is small (less than 30), you would use a t-test to compare the means of the two groups.

Z-Test: Suppose you are a quality control manager at a factory that produces widgets. You want to test whether the mean weight of widgets produced by your factory is the same as the mean weight specified by the manufacturer. You collect a random sample of 100 widgets from the production line and find that the sample mean weight is 10 grams, with a known population standard deviation of 2 grams. Since the sample size is large (greater than 30) and the population standard deviation is known, you would use a z-test to compare the sample mean to the specified mean weight.

**Q2. Differentiate between one-tailed and two-tailed tests.**

**Ans:**
    
A hypothesis test is a statistical test used to determine whether there is enough evidence to reject a null hypothesis. The null hypothesis is a statement that there is no significant difference between the two groups being compared, while the alternative hypothesis is a statement that there is a significant difference.

One-tailed and two-tailed tests refer to the number of tails in the probability distribution used to test the null hypothesis. Here's how they differ:

One-tailed test: A one-tailed test is used when the null hypothesis is rejected only if the sample statistic falls within a certain range on one side of the distribution. In other words, the hypothesis is directional, and the researcher is only interested in detecting a difference in one direction. One-tailed tests are used when there is a strong theoretical or practical reason to expect a difference in a particular direction. For example, a researcher might use a one-tailed test to determine if a new drug is better than an existing drug, rather than just different.

Two-tailed test: A two-tailed test is used when the null hypothesis is rejected if the sample statistic falls within a certain range on either side of the distribution. In other words, the hypothesis is non-directional, and the researcher is interested in detecting a difference in either direction. Two-tailed tests are used when there is no clear theoretical or practical reason to expect a difference in a particular direction. For example, a researcher might use a two-tailed test to determine if there is a significant difference in the mean height of men and women.

**Q3. Explain the concept of Type 1 and Type 2 errors in hypothesis testing. Provide an example scenario for each type of error.**

**Ans:**
    
Type 1 and Type 2 errors are two types of errors that can occur in hypothesis testing.

Type 1 error occurs when the null hypothesis is rejected even though it is true. In other words, it is a false positive result, where the researcher concludes that there is a significant difference when there is no actual difference. The probability of a Type 1 error is denoted by alpha (α), which is the significance level of the test. The smaller the alpha, the less likely a Type 1 error will occur.

Example scenario: Suppose a medical researcher wants to test whether a new drug is more effective than an existing drug. The null hypothesis is that there is no difference between the two drugs, and the alternative hypothesis is that the new drug is more effective. If the researcher rejects the null hypothesis and concludes that the new drug is more effective, but in reality, there is no significant difference between the two drugs, then a Type 1 error has occurred.

Type 2 error occurs when the null hypothesis is not rejected even though it is false. In other words, it is a false negative result, where the researcher concludes that there is no significant difference when there is an actual difference. The probability of a Type 2 error is denoted by beta (β), and it is affected by the sample size, effect size, and significance level of the test. The smaller the beta, the less likely a Type 2 error will occur.

Example scenario: Suppose a researcher wants to test whether a new drug is more effective than an existing drug. The null hypothesis is that there is no difference between the two drugs, and the alternative hypothesis is that the new drug is more effective. If the researcher fails to reject the null hypothesis and concludes that there is no significant difference between the two drugs, but in reality, the new drug is more effective, then a Type 2 error has occurred.

**Q4. Explain Bayes's theorem with an example.**

**Ans:**
    
Bayes's theorem is a mathematical formula that describes the probability of an event based on prior knowledge of conditions that might be related to the event. It is named after Reverend Thomas Bayes, an 18th-century British statistician and Presbyterian minister who first formulated it.

The theorem states that the probability of an event A occurring, given that event B has occurred, can be calculated as:

P(A|B) = P(B|A) x P(A) / P(B)

where:

* P(A|B) is the conditional probability of event A occurring given that event B has occurred
* P(B|A) is the conditional probability of event B occurring given that event A has occurred
* P(A) is the prior probability of event A occurring (i.e., before any new evidence is considered)
* P(B) is the prior probability of event B occurring (i.e., before any new evidence is considered)

Here's an example of how Bayes's theorem can be used:


Suppose there is a factory that produces light bulbs. It is known that 2% of the light bulbs are defective. A quality control inspector checks each bulb and finds that the inspector has a 98% chance of correctly identifying a defective bulb, but also has a 3% chance of falsely identifying a non-defective bulb as defective. If a bulb is randomly selected from the factory and found to be defective, what is the probability that the bulb was actually defective?

Let's define the events as follows:

    A = the bulb is defective

    B = the bulb is identified as defective by the inspector

We know that, 

    P(A) = 0.02 (the prior probability of a bulb being defective)  

    P(B|A) = 0.98 (the probability of the inspector correctly identifying a defective bulb)


We also know that, 

    P(A) = 0.98 (the prior probability of a bulb not being defective)

    P(B|A) = 0.03 (the probability of the inspector falsely identifying a non-defective bulb as defective)

Using the law of total probability, we can find P(B), the probability of the inspector identifying a bulb as defective:

    P(B) = P(A) x P(B|A) + P(A) x P(B|A) = 0.02 x 0.98 + 0.98 x 0.03 = 0.0494

Now we can apply Bayes's theorem:

    P(A|B) = P(B|A) x P(A) / P(B) = 0.98 x 0.02 / 0.0494 = 0.396

So if a bulb is identified as defective by the inspector, the probability that the bulb is actually defective is about 39.6%, which is quite low. This means that there is a significant chance that the bulb is not actually defective and that the inspector made an error. Therefore, additional inspections or tests may be needed to confirm the defectiveness of the bulb.

**Q5. What is a confidence interval? How to calculate the confidence interval, explain with an example.**

**Ans:**
    
A confidence interval is a range of values that is likely to contain a true population parameter with a certain degree of confidence or probability. It is a statistical measure used to estimate the range within which a population parameter (such as a mean or proportion) is likely to fall based on a sample from that population.


Confidence intervals are typically calculated using a formula that takes into account the sample size, standard deviation, and level of confidence desired. The level of confidence is often chosen to be 95%, which means that if the same population were sampled many times, then 95% of the resulting confidence intervals would contain the true population parameter.


Here's an example of how to calculate a confidence interval:

Suppose we want to estimate the average weight of a certain species of bird in a particular region. We take a random sample of 50 birds and measure their weights. We find that the sample mean weight is 250 grams, and the sample standard deviation is 10 grams.


To calculate a 95% confidence interval for the true population mean weight, we can use the following formula:

    CI = X̄ ± Zα/2 * (σ / √n)

where:

    X̄ is the sample mean
    
    Zα/2 is the critical value of the standard normal distribution corresponding to the desired level of confidence (in this case, 95% or 1.96)
    
    σ is the population standard deviation (unknown, so we use the sample standard deviation instead)
    
    n is the sample size
    
Plugging in the values, we get:

    CI = 250 ± 1.96 * (10 / √50) = 250 ± 2.8

So the 95% confidence interval for the true population mean weight of the bird species is (247.2, 252.8) grams. This means that we can be 95% confident that the true population mean weight lies within this range.

**Q6. Use Bayes' Theorem to calculate the probability of an event occurring given prior knowledge of the event's probability and new evidence. Provide a sample problem and solution.**

**Ans:**
    
A company produces two types of products: Type A and Type B. Type A products have a defect rate of 5%, while Type B products have a defect rate of 3%. Suppose a customer buys a product from the company, but does not know which type it is. If the product is found to be defective, what is the probability that it is a Type A product?

Solution:

Let's use Bayes' theorem to calculate the probability that the product is Type A given that it is defective:

    P(A|D) = P(D|A) * P(A) / P(D)

where:

    P(A|D) is the probability that the product is Type A given that it is defective
    
    P(D|A) is the probability that a Type A product is defective = 0.05
    
    P(A) is the prior probability that the product is Type A = 0.5
    
    P(D) is the probability that the product is defective, regardless of its type
    

To calculate P(D), we need to use the law of total probability:

    P(D) = P(D|A) * P(A) + P(D|B) * P(B)

where:

    P(D|B) is the probability that a Type B product is defective = 0.03

    P(B) is the prior probability that the product is Type B = 0.5


Plugging in the values, we get:

    P(D) = 0.05 * 0.5 + 0.03 * 0.5 = 0.04

Therefore, the probability that the product is Type A given that it is defective is:

    P(A|D) = P(D|A) * P(A) / P(D)
    
    = 0.05 * 0.5 / 0.04
    
    = 0.625 or 62.5%

So, given that the product is defective, there is a 62.5% chance that it is a Type A product.

**Q8. What is the margin of error in a confidence interval? How does sample size affect the margin of error? Provide an example of a scenario where a larger sample size would result in a smaller margin of error.**

**Ans:**
    
The margin of error in a confidence interval is the range of values that we add and subtract from the sample mean to create a range of values in which we expect the true population mean to fall within with a certain level of confidence. It is the maximum amount of error we are willing to tolerate when estimating the true population parameter based on the sample statistic.

The margin of error is calculated using the formula:

    Margin of error = Critical value * Standard error

where:

* Critical value is the value from the standard normal distribution or t-distribution that corresponds to the desired level of confidence and sample size

* Standard error is the standard deviation of the sample mean, calculated as the sample standard deviation divided by the square root of the sample size

Sample size affects the margin of error in that a larger sample size will result in a smaller margin of error, assuming all other factors remain the same. This is because a larger sample size reduces the variability of the sample mean, resulting in a more precise estimate of the true population mean.


For example, let's say we want to estimate the average height of all students in a particular school with 95% confidence, and we take two samples of different sizes: one with 50 students and another with 500 students. Assuming the same level of variability in the heights of the students in both samples, the larger sample size will result in a smaller margin of error.

Suppose the sample mean height for the 50-student sample is 165 cm with a sample standard deviation of 5 cm, and the sample mean height for the 500-student sample is also 165 cm with the same sample standard deviation of 5 cm. Using a 95% confidence level and assuming a normal distribution, the critical value for the 50-student sample is 1.96, and for the 500-student sample, it is 1.96 as well.

Calculating the margin of error for each sample size:

    Margin of error for 50-student sample = 1.96 * (5 / √50) ≈ 1.39 cm
    
    Margin of error for 500-student sample = 1.96 * (5 / √500) ≈ 0.22 cm
    

As we can see, the margin of error for the 500-student sample is much smaller than that of the 50-student sample. This means that we can be more confident that the true population mean falls within a narrower range of values with a larger sample size.

**Ans:**
    
To conduct a hypothesis test to determine if the weight loss drug is significantly effective at a 95% confidence level using a t-test, we need to set up the null and alternative hypotheses as follows:

    Null hypothesis (H0): The weight loss drug is not significantly effective, 
    and the sample mean weight loss is not significantly different from zero.

    H0: μ = 0.

    Alternative hypothesis (Ha): The weight loss drug is significantly effective, 
    and the sample mean weight loss is significantly different from zero.

    Ha: μ ≠ 0.


We will use a two-tailed t-test with a significance level of 0.05.

First, we need to calculate the t-statistic using the formula:

    t = (x̄ - μ) / (s / sqrt(n))

where

    x̄ is the sample mean weight loss

    μ is the hypothesized population mean (0 in this case)

    s is the sample standard deviation

    n is the sample size


Plugging in the values, we get:

    t = (6 - 0) / (2.5 / sqrt(50))
      = 18.97

Next, we need to find the critical t-value using a t-distribution table with (n-1) degrees of freedom and a significance level of 0.05. Since the sample size is 50, the degrees of freedom are 49. From the table, we find that the critical t-value is ±2.009.

Since the calculated t-value (18.97) is greater than the critical t-value (±2.009), we can reject the null hypothesis and conclude that the weight loss drug is significantly effective at a 95% confidence level.

Therefore, we can say that there is strong evidence to support that the weight loss drug is effective and that the observed sample mean weight loss of 6 pounds is significantly different from zero.
